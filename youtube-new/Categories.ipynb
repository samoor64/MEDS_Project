```python
import pandas as pd
import json
from pprint import pprint
```

```python
CA_df = pd.read_csv('CAvideos.csv')
```

```python
videos = CA_df.groupby(['title']).sum()
videos
videos_df = videos.reset_index()
videos_df
```

```python
with open('CA_category_id.json') as f:
    data = json.load(f)

pprint(data)
```

```python
#This will hold entire data extracted from json file
find_category={}
# Create a function to automate the task of data collection
for category in data["items"]:
    find_category[category["id"]]=category["snippet"]["title"]
```

```python
records = list(find_category.items())

category_df = pd.DataFrame.from_records(records, columns=['id', 'category'])
print(category_df)
category_df.rename(columns={"id": "category_id"}, inplace=True)
category_df["category_id"] = category_df["category_id"].astype(int)
```

```python
merge_tables = pd.merge(videos_df, category_df, on='category_id', how='outer')
```
